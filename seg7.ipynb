```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_blobs
```

```python
df = pd.read_csv("FLD.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
x = df.iloc[:,[5, 7]].values
x
```

```python
plt.scatter(x[...,0],x[...,1])
plt.xlabel("Ratings")
plt.ylabel("Price")
plt.show()
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    print(wcss)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("No. of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
sns.pairplot(data = df)
```

```python
model = KMeans(n_clusters=3,init="k-means++",random_state=32)
Y_Kmeans = model.fit_predict(x)
Y_Kmeans
```

```python
model.fit(x)
silhouette_score_average = silhouette_score(x, model.predict(x))
silhouette_score_average
```

```python
plt.scatter(x[Y_Kmeans==0,0], x[Y_Kmeans==0,1], label='Cluster 1')
plt.scatter(x[Y_Kmeans==1,0], x[Y_Kmeans==1,1], label='Cluster 2')
plt.scatter(x[Y_Kmeans==2,0], x[Y_Kmeans==2,1], label='Cluster 3')

plt.xlabel("Ratings")
plt.ylabel("Price")
plt.legend()
plt.show()
```
